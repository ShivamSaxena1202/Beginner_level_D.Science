#Missing Value Imputation - Categorical Variable

```python
'''                     NOTE:-
    jb "mode" ka use kr k dataframe me value impute krenge toh uss time dataframe me missing value "Randomly" 
    hona chahiye and 5% of data missing hoga tbhi hm mostly "mode" se missing value impute krte hai.
'''
```

```python
import numpy as np  
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/train.csv")

df.head()
```

```python
# # jb Dataset me kuch limited hi Columns dikh rhe ho toh Use this:- 

# pd.set_option("display.max_columns" , None) # with the help of this command now,seen all columns of This Dataset

# pd.set_option("display.max_rows" , None) # Now Seen all Rows Of this "clean" Dataset
```

```python
# show only "Categorical Variable" into dataframe

cat_vars = df.select_dtypes(include= "object")

cat_vars.head()
```

```python
cat_vars.shape
```

```python
# Konse variable me kitna % data missing hai

miss_val_per = cat_vars.isnull().mean()*100

# miss_val_per = cat_vars.isnull().sum()/cat_vars.shape[0] * 100

miss_val_per
```

```python
# Jiss Column me 20% se jadha Data Missing hai uss ko show krna hai

drop_column = miss_val_per[miss_val_per > 20].keys()
drop_column
```

```python
# drop the column

cat_vars.drop(columns = drop_column , axis=1 , inplace = True)
cat_vars.shape
```

```python
# Data Frame me "categorial variable" k jo jo columns me value missing hai unn ko access krna hai

abc = cat_vars.isnull().mean()*100  
# abc

# ab jaha pr jo "formula" phele variable me use kr k bnaya tha i.e "miss_val_per" tha woh nhi ayega
# kyuki ab hum data me se kuch column "DROP" kr chuke hai.

miss_cat_var = abc[abc > 0].keys()

miss_cat_var
```

```python
miss_cat_var.shape
```

```python
'''
iss saare 11 columns me Missing Value ko impute krna hai :-
['MasVnrType', 
'BsmtQual', 
'BsmtCond', 
'BsmtExposure', 
'BsmtFinType1',
'BsmtFinType2', 
'Electrical', 
'GarageType', 
'GarageFinish',
'GarageQual', 
'GarageCond']

'''
```

```python
# iss column me mannually "missing" value fill krni hai toh

# cat_vars['MasVnrType'].fillna("Missing")
```

```python
# iss column me sbse jadha kosi value hai ye dekhne k liye

cat_vars['MasVnrType'].mode()
```

```python
# iss "column/variable" k count dekhna hai toh like konsi value kitne bar hai:-

cat_vars['MasVnrType'].value_counts()
```

```python
# iss column me missing value fill krna hai :- 'MasVnrType'  

cat_vars['MasVnrType'].fillna(cat_vars['MasVnrType'].mode()[0] , inplace= True)
```

```python
cat_vars['MasVnrType'].value_counts()
```

```python
# check in "MasVnrType" how many null value are present

cat_vars['MasVnrType'].isnull().sum()
```

```python
cat_vars.isnull().sum()
```

```python
cat_vars.isnull().sum().sum()
```

```python
# saare "columns/variables" me unn ki mode value fill krni hai (for-loop) ki help se

cat_vars_copy= cat_vars.copy()

for var in miss_cat_var:
    cat_vars_copy[var].fillna(cat_vars[var].mode()[0] , inplace = True)
    print(var , "=" , cat_vars[var].mode()[0])
```

```python
cat_vars_copy.isnull().sum()
```

```python
cat_vars_copy.isnull().sum().sum()
```

#Data Distribution

```python
plt.figure(figsize=(16,9))
for i,var in enumerate(miss_cat_var):
    plt.subplot(4,3,i+1)
    plt.hist(cat_vars_copy[var],label="Impute")
    plt.hist(cat_vars[var].dropna(),label="Original")
    plt.legend()
```

```python
df.update(cat_vars_copy)
df.drop(columns=drop_column,inplace=True)
```
