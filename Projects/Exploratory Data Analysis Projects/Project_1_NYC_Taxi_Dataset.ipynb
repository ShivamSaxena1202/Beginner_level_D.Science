```python
# change datatype of (pickup_datetime & dropoff_datetime) Column/Variable  into dataset

data['pickup_datetime'] = pd.to_datetime(data['pickup_datetime'])

data["dropoff_datetime"] = pd.to_datetime(data["dropoff_datetime"])
```

```python
# find (pickup_day and drop_day) and add this two more columns/Variable  into main dataset 

data["pickup_day"] = data["pickup_datetime"].dt.day_name()

data["dropoff_day"] = data["dropoff_datetime"].dt.day_name()
```

```python
# find (pickup_month and drop_month) and add this two more columns/Variable into main dataset

data["pickup_month"] = pd.DatetimeIndex(data["pickup_datetime"]).month

data["dropoff_month"] = pd.DatetimeIndex(data["dropoff_datetime"]).month
```

# Data Visualization

```python
# check "passenger_count" Variable data_distribution

dv = sns.countplot(x = data['passenger_count'])
plt.title("Dist of Passenger count")
```

```python
# Want to add number of passengers on "pillor" ,means How many number of passengers are in each pillor

dv = sns.countplot(x = data['passenger_count'])
plt.title("Dist of Passenger count")

for p in dv.patches:
    height = p.get_height()
    dv.text(x = p.get_x() + (p.get_width()/2),
    y = height + 0.2  , ha = 'center' , s = '{:.0f}'.format(height))
    
plt.show()
```

```python
data.head()
```

```python
# which "day" are very high/Rush in full week  

figure, ax = plt.subplots(nrows = 1, ncols=2, figsize = (15,5))
sns.countplot(x = 'pickup_day', data = data, ax = ax[0])
ax[0].set_title(' no of pickups done on each day of the week')

sns.countplot(x = 'dropoff_day', data = data, ax = ax[1])
ax[1].set_title(' no of dropoffs done on each day of the week')

plt.tight_layout()
```

```python
# check "trip_duration" column/Variable's data_distribution

x = data["trip_duration"]

sns.distplot(x)
```

```python
# dataset me ek aur column add krenge jo btayega ki passenger ko (drop and Pick) krne ka time kya hai
# like morning,evening,etc-etc

def timezone(x):
#     Allocate time zone to each row
    if x>=datetime.time(4, 0, 1) and x <=datetime.time(10, 0, 0):    # (Hour,min,second)
        return 'morning'
    elif x>=datetime.time(10, 0, 1) and x <=datetime.time(16, 0, 0):
        return 'midday'
    elif x>=datetime.time(16, 0, 1) and x <=datetime.time(22, 0, 0):
        return 'evening'
    elif x>=datetime.time(22, 0, 1) or x <=datetime.time(4, 0, 0):
        return 'late night'
    
data['pickup_timezone']=data['pickup_datetime'].apply(lambda x :timezone(datetime.datetime.strptime(str(x), "%Y-%m-%d %H:%M:%S").time()) )
data['dropoff_timezone']=data['dropoff_datetime'].apply(lambda x :timezone(datetime.datetime.strptime(str(x), "%Y-%m-%d %H:%M:%S").time()) )
```

```python
# 'Distribution of pickup_timezone'

ax = sns.countplot(x=data['pickup_timezone']);
plt.title('Distribution of pickup_timezone')
for p in ax.patches:
    height = p.get_height()
    ax.text(x = p.get_x()+(p.get_width()/2), # x-coordinate position of data label, padded to be in the middle of the bar
    y = height+0.2, ha = 'center',s = '{:.0f}'.format(height)) # data label, formatted to ignore decimals
    #ha = ‘center’) # sets horizontal alignment (ha) to center
plt.xticks(rotation = 'vertical')    
plt.show()
```

```python
#  'Distribution of dropoff_timezone

ax = sns.countplot(x=data['dropoff_timezone']);
plt.title('Distribution of dropoff_timezone')
for p in ax.patches:
    height = p.get_height()
    ax.text(x = p.get_x()+(p.get_width()/2), # x-coordinate position of data label, padded to be in the middle of the bar
    y = height+0.2, ha = 'center',s = '{:.0f}'.format(height)) # data label, formatted to ignore decimals
    #ha = ‘center’) # sets horizontal alignment (ha) to center
plt.xticks(rotation = 'vertical')    
plt.show()
```

```python
# Data Distribution of How "Hour of the day" are work

figure, ax = plt.subplots(nrows = 1, ncols=2, figsize = (15,5))

data['pickup_hour'] = data['pickup_datetime'].dt.hour
data.pickup_hour.hist(bins = 24 , ax = ax[0])
ax[0].set_title('dist of pickups hrs')

data['dropoff_hour'] = data['dropoff_datetime'].dt.hour
data.dropoff_hour.hist(bins = 24 , ax = ax[1])
ax[1].set_title('dist of pickups hrs')

plt.tight_layout()
```
